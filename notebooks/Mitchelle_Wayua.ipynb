```python
#Geospatial scatter plot to map accidents
import folium

# Drop rows with missing latitude or longitude
df_cleaned = df.dropna(subset=['tb_latitude', 'tb_longitude'])

# Create a basic map centered at the average latitude and longitude
m = folium.Map(location=[df_cleaned['tb_latitude'].mean(), df_cleaned['tb_longitude'].mean()], zoom_start=12)

# Add points to the map
for _, row in df_cleaned.iterrows():
    folium.CircleMarker(
        location=(row['tb_latitude'], row['tb_longitude']),
        radius=3, color='red', fill=True
    ).add_to(m)

# Save the map to an HTML file
m.save('traffic_incidents_map.html')
print("Map successfully created!")
```

```python
#Count of incidents based on party at fault
# Group data by 'party_at_fault' and 'collision_severity' to create a breakdown
party_severity = df.groupby(['party_at_fault', 'collision_severity']).size().unstack(fill_value=0)

# Normalize to show proportions instead of raw counts
party_severity_normalized = party_severity.div(party_severity.sum(axis=1), axis=0)

# Plot the stacked bar chart
party_severity_normalized.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

# Add title, axis labels, and legend
plt.title('Proportion of Collision Severity by Party at Fault', fontsize=14)
plt.xlabel('Party at Fault', fontsize=12)
plt.ylabel('Proportion', fontsize=12)
plt.legend(title='Collision Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
```

## "Injury (Complaint of Pain)" is the most common outcome across all parties at fault: 
The green sections dominate the bars for all categories, suggesting that minor injuries are by far the most frequent collision outcome, regardless of which party is responsible.

## "Injury (Other Visible)" is the second most frequent severity: 
Represented by brown, this category consistently appears in all bars, though in smaller proportions compared to "Complaint of Pain." This indicates visible but not severe injuries are fairly common.

## Severe injuries, fatalities, and medical incidents are rare: 
The gray ("Injury (Severe)"), blue ("Fatal"), and cyan ("Medical") segments are noticeably small, reflecting that while these outcomes do occur, they make up a small proportion of total collisions.

## Consistency across fault categories: 
The proportions of collision severities remain relatively consistent across all "Party at Fault" categories (1.0 through 5.0), implying no single party type is disproportionately associated with severe outcomes.

This visualization highlights that the majority of traffic collisions result in less severe injuries, while severe incidents and fatalities are outliers.
