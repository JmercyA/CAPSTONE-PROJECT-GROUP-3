```python
import pandas as pd
df=pd.read_csv('Traffic_Crashes_Resulting_in_Injury_20250307.csv')
df
```

```python
df.columns
```

```python
df_selected=df[[
       'collision_datetime', 'collision_date', 'collision_time',
       'accident_year', 'month', 'day_of_week', 'time_cat',
       'secondary_rd', 'distance', 'direction', 'weather_1', 
       'collision_severity', 'type_of_collision', 'mviw', 'ped_action',
       'road_surface', 'road_cond_1', 'road_cond_2', 'lighting',
       'control_device', 'intersection', 'vz_pcf_code',  'number_killed', 'number_injured',
       'party_at_fault', 'party1_type', 'party1_dir_of_travel',
       'party1_move_pre_acc', 'party2_type', 'party2_dir_of_travel',
       'party2_move_pre_acc']]
```

```python
df_selected.head()
```

```python
df_selected.isna().sum()
```

```python
df_selected['collision_time'].fillna(method='ffill',inplace=True)
df_selected
```

```python
df_selected.isna().sum()
```

```python
import numpy as nd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_selected['collision_time'].value_counts()
```

```python
df_selected['collision_time'].value_counts()[0:20].plot(kind='bar')
plt.title('collision_time')
plt.xlabel('collision_time')
plt.ylabel("frequency")
```

```python
df_selected['month'].value_counts()[0:20].plot(kind='bar')
plt.title('month')
plt.xlabel('month')
plt.ylabel("frequency")
```

```python
df_selected['day_of_week'].fillna(method='ffill',inplace=True)
df_selected
```

```python
df_selected.isna().sum()
```

```python
df_selected['day_of_week'].value_counts()[0:20].plot(kind='bar')
plt.title('day_of_week')
plt.xlabel('day_of_week')
plt.ylabel("frequency")
```

```python
day_severity_count=df_selected.groupby(['day_of_week','collision_severity']).size().reset_index(name='count')
day_severity_count
```

```python
pvt_table=pd.pivot_table(day_severity_count,values='count',index=['day_of_week'],columns=['collision_severity'],aggfunc='sum')
pvt_table
```

```python
pvt_table=pvt_table.reset_index().rename_axis(None,axis=1)
pvt_table
```

```python
colors = ['green', 'lightgreen', 'blue', 'lightblue', 'red'] ## Colors for the bars
pvt_table.set_index('day_of_week').plot(kind='bar', stacked=True, color=colors) ## Plot
plt.ticklabel_format(style='plain', useOffset=False, axis='y') ## No offset
plt.gca().set_ylabel("Count") ## Set Y-axis
```
